```python
types = df["alias"].tolist()
uniqueTypes = []
for i in types:
  for j in i:
    if j not in uniqueTypes:
      uniqueTypes.append(j)

foo = []
for k in uniqueTypes:
  bar = []
  for l in range(0,len(df)):
    if k in df['alias'].iloc[l]:
      bar.append(1)
    else:
      bar.append(0)
  foo.append(bar)

count = 0
for i in uniqueTypes:
  df[i] = foo[count]
  count += 1
df.drop(['image_url', 'url', 'transactions'], axis=1, inplace = True)
df
```

```python
# Running Kmodes clustering and determining which number of clusters minimizes "cost", without including too many clusters
cost = []
K = range(1,30)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(df.iloc[:,19:])
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```
